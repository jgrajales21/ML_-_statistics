If we want to scrape tags with more specificity we use the find function and then add the keyword associated with a property. In this case we want to get the table data ('td') with class valued as 'column-3'.

```python
sample_row.find('td',class_='column-3').get_text()
```

Now we do this for all of the data on the page.

```python
url = 'https://thefactfile.org/countries-currencies-symbols/'
import requests
from bs4 import BeautifulSoup
response = requests.get(url)
page_text = response.text
soup = BeautifulSoup(page_text)
# tr = table row
all_rows = soup.find_all('tr')
for row in all_rows:
    name = row.find('td', class_='column-3')
    iso = row.find('td', class_='column-4')
    print(name,iso)
```

Now we request for the html corresponding to our url. Recall that under-the-hood we are connecting to some server, asking for the html correspding to the url, and then using a compiler on our machine to run the html and access the information on the website.

```python
def get_currency_list():
    currency_list = list()
    url = 'https://thefactfile.org/countries-currencies-symbols/'
    # requests will run the required protocols to get the url's html
    import requests
    # check if we successfully received the data 
    response = requests.get(url)
    # a response code of 200 signifies that we successfully contacted and received data from the url's server
    if not response.status_code == 200:
        return currency_list
    return currency_list 
```
